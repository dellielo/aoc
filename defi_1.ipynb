```python
import re
with open('./data/1.txt') as fic :
    lines = [line for line in fic]

dico = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9}
sum = 0
def find_digit(txt):
    matches = re.finditer("(?:one|two|three|four|five|six|seven|eight|nine|\d)", txt)
    m = [match for match in matches]
    return m

for line in lines :
    m = find_digit(line)
    ind = [int(dico.get(match.group(),match.group())) for match in m]
    first = ind[0]
    sub = line[m[-1].start()+1: -1]

    m_sub = find_digit(sub)
    if m_sub : 
        # print("sub!", sub)
        ind = [int(dico.get(match.group(),match.group())) for match in m_sub]
        # print("sub!!", ind)
    last = ind[-1]
    
    if m[-1].end() < len(line) :
        if m_sub :
            if m_sub[-1].end() < len(line):
                print(line, first ,last)
        print(line, first ,last)
    sum += (first*10 + last)

print(sum)
```
